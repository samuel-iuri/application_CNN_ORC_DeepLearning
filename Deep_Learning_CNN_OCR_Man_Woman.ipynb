## Criando a Rede Neural Covolucional

```python
# Criando a rede neural convolucional com uma estrutura pré-treinada
Modelo_Basico = ResNet50(weights = 'imagenet', include_top = False,
                      input_tensor = Input(shape = (256,256,3)))  #PADRÃO KERAS PARA TAMANHO DE IMAGEM:  (224, 224, 3)
```

```python
# Verificando os dados da Rede Neural
Modelo_Basico.summary()
```

```python
# Verificando as camadas da Rede
len(Modelo_Basico.layers)
```

```python
# Congelando os pesos da Rede Neural
# Vamos deixar as ultimas camdas livres para serem treinadas
for Camada in Modelo_Basico.layers[:-10]:
    Camada.trainable = False
```

```python
# Definindo a camada de inicio
Modelo_Inicio = Modelo_Basico.output
Modelo_Inicio = AveragePooling2D()(Modelo_Inicio)
Modelo_Inicio = Flatten()(Modelo_Inicio)
Modelo_Inicio = Dense(256, activation = 'relu')(Modelo_Inicio) #128 neurônios ou 256
Modelo_Inicio = Dropout(0.2)(Modelo_Inicio)
Modelo_Inicio = Dense(256, activation = 'relu')(Modelo_Inicio) #128 neurônios ou 256
Modelo_Inicio = Dropout(0.2)(Modelo_Inicio)
# Se forem 2 classes usamos a função 'sigmoide',
# para classificar mais de 2 classes usamos 'softmax'
Modelo_Inicio = Dense(2, activation = 'sigmoid')(Modelo_Inicio)
```

```python
# Criando o Modelo da Rede Neural
Modelo = Model(inputs = Modelo_Basico.input, outputs = Modelo_Inicio)
```

```python
# Configurando alguns paramentros da Rede
Modelo.compile(
    # Função de Erro
    loss = 'binary_crossentropy', #'categorical_crossentropy'

    # Algoritimo de ajustes dos Pesos
    optimizer = optimizers.RMSprop(learning_rate = 1e-4, decay = 1e-6),

    # Metrica de avaliação
    metrics = ['accuracy']
)
```

```python
# Salvar os dados da Rede Neural
Salvando_Rede = ModelCheckpoint(filepath='PESOS_RNC_5.hdf5',
                                save_best_only = False)
#True para não salvar os pesos
```

```python
# Função para acessar as imagens do diretorio e preparar as imagens

Treino_Generator = Image_Generator.flow_from_directory(
                                                       batch_size = 8,
                                                       directory = Diretory_Images_Train,
                                                       shuffle = True,
                                                       target_size = (256, 256),
                                                       class_mode = 'categorical',
                                                       subset = 'training')
```

```python
#CUIDADO AO EXECUTAR! GRANDE TEMPO DE APRENDIZAGEM DO MODELO

# Rodar o Modelo
Historico = Modelo.fit_generator(Treino_Generator, epochs = 60, callbacks = [Salvando_Rede])
```

```python
# Acessando os dados da Rede
Historico.history.keys()
```

```python
# Plotar a Acuracia do Modelo

# Definindo uma figura
plt.figure( figsize=(10,6) )
# Plotando a Acuracia
plt.plot( Historico.history['accuracy'] )
# Plotando o erro
plt.plot( Historico.history['loss'] )
# Definindo o Titulo
plt.title('Erro e Acerto da Rede')
# Definindo os Labels
plt.xlabel('Épocas')
plt.ylabel('Acerto x Erro')
# Definindo a Legenda
plt.legend(['Acerto', 'Erro']);
```

```python
prediction = []
original = []
image = []

for i in os.listdir(Directory_Images_Test):
    for item in os.listdir(os.path.join(Directory_Images_Test, str(i))):
        string = os.path.join(Directory_Images_Test , str(i) , item)
        img = cv2.imread(string)
        #print(img)
        img = cv2.resize(img, (256, 256))
        image.append(img)
        img = img/255
        img = img.reshape(-1, 256, 256, 3)
        predict = Modelo.predict(img)
        predict = np.argmax(predict)
        if(predict == 0):
            prediction.append('Female')
        if(predict == 1):
            prediction.append('Male')
        original.append(i)
```
